```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
accuracy = [0.9649, 0.9963, 0.9968, 0.9970, 0.9971, 0.9971, 0.9971]
epochs = [1, 2, 3, 4, 5, 6, 7]
loss = [0.1209, 0.0106, 0.0093, 0.0089, 0.0085, 0.0085, 0.0086]
val_accuracy = [0.9962, 0.9967, 0.9970, 0.9968, 0.9972, 0.9969, 0.9972]
val_loss = [0.0108, 0.0090, 0.0090, 0.0090, 0.0081, 0.0090, 0.0082]
```

```python
def plot(x, y, label):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=label)
    plt.title(f"Graph of {label}")
    plt.xlabel("Epochs")
    plt.ylabel(label)
    plt.show()
```

```python
plot(epochs, accuracy, "train_accuracy")
```

```python
plot(epochs, loss, "train_loss")
```

```python
plot(epochs, val_accuracy, "val_accuracy")
```

```python
plot(epochs, val_loss, "val_loss")
```
